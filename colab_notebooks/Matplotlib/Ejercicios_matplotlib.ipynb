## Ejercicio 4

Crea un array con 100 valores entre 0 y 20 (ambos incluidos) distribuidos de forma homogénea.

Utiliza los valores de este array para el eje de abcisas.

Representa la función $y = 2x$



```python
# Aquí estamos indicando como queremos que trabaje el notebook con las graficas
# En este caso vamos a trabajar con graficas estaticas.

%matplotlib inline

# Creamos el array para el eje de abcisas - x

abcisas_array = np.arange(0, 20, step = 0.25)
print(abcisas_array, "\n")

# Hacemos el calculo para el eje de ordenadas - y

sort_array = 2 * abcisas_array
print(sort_array, "\n")

# Hacemos la representación grafica

plt.plot(abcisas_array, sort_array)
```

## Ejercicio 5

Usa el mismo array para mostrar en una gráfica tanto el seno como el coseno.

```python
# Creamos los para el seno .sin como para el coseno .cos

line_seno = np.sin(abcisas_array)
line_coseno = np.cos(abcisas_array)

# Ahora montamos la grafica con las dos lineas

# Para el seno .sin:

plt.plot(abcisas_array, line_seno, '-')

# Para el coseno .cos:

plt.plot(abcisas_array, line_coseno, '--')
```

## Ejercicio 6

Igual que el ejercicio anterior pero mostrando dos gráficas diferentes, una para el seno y otra para el coseno.

```python
"""

  Con el metodo subplot estamos configurando paneles diverson para las graficas

  Por ejemplo:
    .subplot(1, 1, 1) estamos indicando lo siguiente:
        - En el primer argumento estamos indicando el numero de filas que va
          a tener el panel donde se alojaran las graficas, este numero hay que
          ir incrementandolo según el numero de graficas que queramos repre-
          sentar.
        - En el segundo argumento estamos indicando el numero de columnas, su
          funcionamiento es el mismo que el argumento 1, pero esta representa 
          las graficas en columnas.
        - En el tercer argumento estamos indicando en que panel queremos
          que se represente la grafica, este numero puede configurarse como 
          queramos, solo cambiara la posición de visualización de la grafica.

"""
plt.subplot(3, 1, 1) #(rows, columns, panel number)

plt.plot(abcisas_array, line_seno)

plt.subplot(3, 1, 2)

plt.plot(abcisas_array, line_coseno)

plt.subplot(3, 1, 3)

plt.plot(abcisas_array, line_seno, '-')
plt.plot(abcisas_array, line_coseno, '--')
```

## Ejercicio 7

Crea dos arrays de 20 elementos cada uno. El primero con los números del 0 al 9 y el segundo con los cuadrados de los primeros.

Realiza una gráfica cuyos datos para la $x$ sea el primer array y para la $y$, el segundo.

```python
# Creamos los nuevos arrays

array_1 = np.arange(0, 9, step = 0.5, dtype = np.float16)
array_2 = array_1 ** 2

plt.xlabel("Array_1")
plt.ylabel("Array_2")

plt.plot(array_1, array_2)
```

## Ejercicio 8

Muestra en una misma gráfica, las funciones lineal, cuadrática y cúbica.

Debe aparecer una leyenda indicando a qué corresponde cada línea.

La gráfica debe tener un título.

Las líneas deben ser cada una de un tipo: continua, con pequeñas líneas discontinuas y con puntos.

```python
# Primero vamos a crear un array para la linea de cubic

array_3 = array_1 ** 3

# Ahora nombramos las lineas y le ponemos titulo a la grafica

plt.xlabel("x label")
plt.ylabel("y label")
plt.title("Ejercicio 8")

"""

    Para generar la leyenda debemos de introducir como argumento 
    en .plot el atributo label y despues llamar al metodo .legend()
    
"""

plt.plot(array_1, array_1, '-', label = "linear")
plt.plot(array_1, array_2, '--',label = "quadratic") 
plt.plot(array_1, array_3, '-.', label = "cubic")


# Llamamos al metodo para mostrar la leyenda en la grafica

plt.legend()
```
